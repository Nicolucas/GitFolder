```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

```python
t=np.loadtxt(open('monthrg.dat.txt','r'))
```

```python
#se filtran los datos en diferentes vectores.
year=t[:,0]
month=t[:,1]
numTomas=t[:,2]
Manchas=t[:,3]
desviacion=t[:,4]
dt=1
#numero de datos
size=shape(Manchas)[0]
size
```

```python
plt.plot(Manchas)
```

```python
from scipy.fftpack import fft, fftfreq
fft_x = fft(Manchas) / size # FFT Normalized
freq = fftfreq(size, dt) # Recuperamos las frecuencias
plt.plot(freq,np.abs(fft_x),'k,')
```

```python
posicion=0
for i in (range(size)):
    dato=Manchas[i]
    if (dato==-99):
        posicion=i+1
posicion
#Tengo la posicion donde ya dejo de perder datos
```

```python
novo=np.empty(0);
for i in range(posicion,size):
    novo = np.append(novo,Manchas[i])
plt.plot(novo)
#A partir de aca tiene sentido la eleccion de datos en la grafica
numDatos=shape(novo)[0]
numDatos
```

```python
from scipy.fftpack import fft, fftfreq
fft_x = fft(novo) / numDatos # FFT Normalized
freq = fftfreq(numDatos, dt) # Recuperamos las frecuencias
plt.plot(freq,np.abs(fft_x),'k,')
#se tiene la frecuencia de los datos utiles
```

```python
minimo=freq[1]
pos=0
for i in (range(1,numDatos)):
    if (freq[i]>minimo):
        minimo=freq[i];
        pos=i;
pos
```

```python
novo1=np.empty(0);
for i in range(1,numDatos):
    data=freq[i]-minimo;
    novo1 = np.append(novo1,data);
novo1
```
