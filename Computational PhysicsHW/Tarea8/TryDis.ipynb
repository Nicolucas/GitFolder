## RungeKutta:

```python
#definicion de la funcion p' 
def p_prime(p,t,alpha,beta,s0,km1,km2):
    p0=alpha*(s0-p)
    p1=km1+s0-p
    p2=(beta*p)/(km2+p)
    return (p0/p1)-p2


#pasos del RungeKutta
h=0.1
n_punt=int(((t[-1]-t[0])+h)/h)
print n_punt
```

```python
#definicion del metodo RungeKutta
def paso_runge_kutta(p,t,alpha,beta,s0,km1,km2):
    
    k1=p_prime(p,t,alpha,beta,s0,km1,km2)
    
    #Primer paso
    t_t=t+(h/2.0)
    p_t=p+(h/2.0*k1)
    k2=p_prime(p_t,t_t,alpha,beta,s0,km1,km2)
    
    #Segundo
    p_t=p+(h/2.0*k2)
    k3=p_prime(p_t,t_t,alpha,beta,s0,km1,km2)
    
    #Tercero
    t_t=t+h
    p_t=p+(h*k3)
    k4=p_prime(p_t,t_t,alpha,beta,s0,km1,km2)
    
    #Cuarto
    k_ave=(k1+(2.0*k2)+(2.0*k3)+k4)/6.0
    
    t_new = t + h
    p_new = p+(h*k_ave) 
    return p_new
```

### ------------------------------------------------------------------------------------------

### $Modelo$ $a$ $usar:$ Definicion de la aplicacion de RungeKutta, es decir el modelo de la funcion

Definicion del modelo de la ecuacion a usar

```python
def simular(alpha,beta,s0,km1,km2):
    t_r=linspace(t[0],t[-1],n_punt)
    p_r=zeros(n_punt)
    p_r[0]=p[0]
    
    for i in range(n_punt-1):
        p_r[i+1]=paso_runge_kutta(p_r[i],t_r[i],alpha,beta,s0,km1,km2)
    return t_r,p_r
```

filtro de los datos a usar, usando los datos dados por el RungeKutta en el paso anterior

```python
def modelo(alpha1,beta1,s01,km11,km21):
    
    #print (alpha1,beta1,s01,km11,km21)
    
    t_r,p_r=simular(alpha1,beta1,s01,km11,km21)
    
    rta=zeros(numData)
    for ti in range(len(t)):
        for ala in range (len(t_r)):
            if t[ti]==t_r[ala]:
                rta[ti]=p_r[ala]
    return rta
    #print rta
                
```

### Definicion del $X²$

Se define una funcion que encuentre el $X^2$ de los valores del modelo y los reales.

```python
def buscado(y_model):
    #print type(y_model)
    chi_squared=0.0
    for i in range(numData):
        chi_squared += (p[i]-y_model[i])**2
    return chi_squared
```
